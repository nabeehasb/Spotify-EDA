```python
##Top 5 loudest tracks
```

```python
top_five_artist_danceable_song= df[["danceability","song_title","artist"]].sort_values(by="danceability",ascending=False)[:5]
top_five_artist_danceable_song
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability",y="song_title", data=top_five_artist_danceable_song)
plt.title("Top 5 danceable Tracks")
plt.savefig("dance.jpg")
plt.show()
```

```python
###Top 10 Instrumentalness tracks
```

```python
top_ten_instrumental_tracks= df[["instrumentalness","song_title","artist"]].sort_values(by="instrumentalness",ascending=False)[:5]
top_ten_instrumental_tracks
```

```python
plt.figure(figsize=(12, 7))
plt.pie(
    x=top_ten_instrumental_tracks["instrumentalness"],  # Using actual values
    labels=top_ten_instrumental_tracks["song_title"],   # Using song titles as labels
    autopct='%1.2f%%', 
    startangle=140  # Optional: Adjust the start angle for better visualization
)

plt.title("Top 10 Instrumental Tracks by Instrumentalness")  # Adding a title
plt.savefig("TEN.jpg")
plt.show()
```

```python
##Multiple Feature plots
```

```python
df.columns
```

```python
interest_feature_cols=["tempo","loudness","acoustioness","danceability","duration_ms","energy","instrumentalness","liveness","speechiness","valence"]
```

```python
for feature_col in interest_feature_cols:
    if feature_col not in df.columns:
        print(f"Column '{feature_col}' not found in DataFrame. Skipping...")
        continue  # Skip missing columns

    pos_data = df[df["target"] == 1][feature_col]
    neg_data = df[df["target"] == 0][feature_col]

    plt.figure(figsize=(12, 7))

    sns.histplot(pos_data, 
                 bins=30, 
                 alpha=0.7, 
                 label="Positive", 
                 color="green", 
                 kde=True)

    sns.histplot(neg_data, 
                 bins=30, 
                 alpha=0.7, 
                 label="Negative", 
                 color="red", 
                 kde=True)

    plt.legend(loc="upper right")
    plt.title(f"Positive and Negative Histogram Plots for {feature_col}")
    plt.xlabel(feature_col)
    plt.ylabel("Count")
    plt.savefig("Multi.jpg")
    plt.show()
```
